```python
import torch
import torch.nn as nn
import numpy as np
import plotly.offline as py
import plotly.graph_objects as go
py.init_notebook_mode(connected=True)
```

```python
x = torch.tensor([7, 1], dtype = torch.float, requires_grad = True)
def param():
    yield x

#optim = torch.optim.SGD(param(), lr = 1)
#optim = torch.optim.Adadelta(param())
optim = torch.optim.Adam(param(), lr = 0.9)
#optim = torch.optim.ASGD(param(), lr = 1)
steps = 100
x_step = np.zeros((steps+1, 2), dtype = np.float32)
x_step[0] = x.detach().numpy()
for i in range(steps):
    #y = x[0]*x[0] + x[1]*x[1]
    y = torch.sin(x[0]/8) + torch.cos(x[1]/2)
    optim.zero_grad()
    y.backward()
    optim.step()
    x_step[i+1] = x.detach().numpy()
```

```python
x_contour = []
y_contour = []
z_contour = []
for _y in np.linspace(-20, 20, 100):
    for _x in np.linspace(-20, 20, 100):
        #z.append(_x*_x + _y*_y)
        z_contour.append(np.sin(_x/8) + np.cos(_y/2))
        x_contour.append(_x)
        y_contour.append(_y)
```

```python
fig = go.Figure()
fig.add_trace(go.Contour(x = x_contour, y = y_contour, z = z_contour))
fig.add_trace(go.Scatter(x = x_step.transpose()[0], y = x_step.transpose()[1], mode = 'markers'))
fig.show()
```
